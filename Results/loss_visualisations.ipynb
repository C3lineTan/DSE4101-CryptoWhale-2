```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "../Results/loss_results.csv"
df = pd.read_csv(path)
```

```python
# Drop Random Synthesiser
df = df[~df["models"].str.contains("Random", case=False)].copy()

# Helper for plotting
def plot_loss(df_in, col, title, xlab):
    d = df_in[["models", col]].sort_values(col, ascending=True)
    plt.figure(figsize=(7, 4))
    bars = plt.barh(d["models"], d[col], edgecolor="black", color="skyblue")

    # Label bars
    for bar, val in zip(bars, d[col].values):
        plt.text(val, bar.get_y() + bar.get_height()/2,
                 f"{val:.6f}", va="center", ha="left", fontsize=9)

    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel("Model")
    plt.grid(axis="x", alpha=0.25)
    plt.tight_layout()
    plt.show()
```

```python
# === BTC charts ===
plot_loss(df, "btc_rmse",  "BTC: RMSE",  "RMSE")
plot_loss(df, "btc_qlike", "BTC: QLIKE", "QLIKE")
```

```python
# === ETH charts ===
plot_loss(df, "eth_rmse",  "ETH: RMSE",  "RMSE")
plot_loss(df, "eth_qlike", "ETH: QLIKE", "QLIKE")
```
